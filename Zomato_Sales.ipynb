````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df=pd.read_csv("/content/Zomato data .csv")
````

````python
df.head()
````

***DATA CLEANING***

````python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
df['rate']=df['rate'].apply(handleRate)
print(df.head())
````

````python
df.info
````

**Type of Resturants**

````python
df.head()
````

````python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('Type of Resturants')
````

**Majority Customer orders from Dining type of resturants**

````python
df.head()
````



```
Votes



````python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data}) # Changed pd.df to pd.DataFrame
plt.plot(result,c="green", marker="o")
plt.xlabel("Type of Resturant" , c="red", size=20)
plt.ylabel("votes", c="red", size=20)
````

**Dining has received maximum votes**

````python
plt.hist(df['rate'],bins=10)
plt.title("ratings distribution")
plt.show
````

**Majority resturant receives rating from 3.5 to 4**

````python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate', data=df)
````

Conclusion: Offline orders receives lower rating in comparison to online order

````python
pivot_table=df.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d') # Changed "Y1GnBu" to "YlGnBu"
plt.title("Heatmap")
plt.xlabel("Online order")
plt.ylabel("Listed in (Type)")
plt.show()
````

Conclusion : Client prefer to place orders in resturants, but prefer online orders from cafes
